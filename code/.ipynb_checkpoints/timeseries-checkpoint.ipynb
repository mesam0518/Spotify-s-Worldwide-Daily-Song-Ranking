```python
#Set a rolling mean with windows of 1, 15, and 30.
rolmean1  = streams.rolling(window = 1).mean()
rolmean4  = streams.rolling(window = 7).mean()
rolmean13 = streams.rolling(window = 15).mean()

# Plot
plt.figure(figsize=(16,5))
plt.plot(rolmean1, color='blue', label='Rolling Mean (1 day)')
plt.plot(rolmean4, color='orange', label='Rolling Mean (5 days)')
plt.plot(rolmean13, color='grey', label='Rolling Mean (15 days)')

plt.legend(loc='best')

plt.title('Rolling Mean for 1, 15 and 30 Months')
plt.show()
```

```python
print(f'Autocorrelation between streams and 1-lag streams: {round(streams["streams"].autocorr(lag=1),4)}')
print(f'Autocorrelation between streams and 15-lag streams: {round(streams["streams"].autocorr(lag=7),4)}')
print(f'Autocorrelation between streams and 20-lag streams: {round(streams["streams"].autocorr(lag=15),4)}')
print(f'Autocorrelation between streams and 30-lag streams: {round(streams["streams"].autocorr(lag=30),4)}')
```

```python
#365 days  Autocorr and Partial Autocorr
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(211)
fig = plot_acf(streams, lags=50, ax=ax1)
ax2 = fig.add_subplot(212)
fig = plot_pacf(streams, lags=50, ax=ax2);

# If you do not use plt.show() it will plot the same vizualization twice.
# This may be a bug you can fix and submit to contribute to open source!
```

```python
streams['lag_1'] = streams['streams'].shift(1)

# Create a column called `lag_2` that lags streams by 6 days.
streams['lag_2'] = streams['streams'].shift(6)

# Create a column called `seasonal_12` that lags streams by one month.
streams['month_30'] = streams['streams'].shift(30)

# Create a variable called `time` that takes on a value of 0 in January 1st 2019,
# then increases by 1 each day until the end of the dataframe.
streams['time'] = range(0, streams.shape[0])
```
